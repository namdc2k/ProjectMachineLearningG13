```python
data_df=pd.read_csv("weather.csv")
```

```python
data_df.head()
```

```python
data_df=data_df.replace(regex=['-',':00','°c','mm','%','mb','km/h'], value="")
data_df
```

```python
int_colums=['day','time']
float_colums=['tem','rain','rain%','cloud','pressure','wind','gust']
data_df[int_colums] = data_df[int_colums].astype(int)
data_df[float_colums] = data_df[float_colums].astype(float)
data_df
```

```python
data_df.info()
```
